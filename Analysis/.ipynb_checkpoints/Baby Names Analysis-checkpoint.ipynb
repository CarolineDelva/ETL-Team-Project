```python
#Importing dependencies
import pandas as pd 
from sqlalchemy import create_engine
```

```python
#Creating file path for csv
uk_boys_names_file = "../Datasets/UK_Boys_2011.csv"
```

```python
#Reading csv and printing the head
uk_boys_names_df = pd.read_csv(uk_boys_names_file, encoding='utf-8')
uk_boys_names_df.head()
```

```python
#Selecting the desired columns
new_uk_boys_names_df = uk_boys_names_df[['Rank', 'Name', 'Count']].copy()
new_uk_boys_names_df.head()
```

```python
#Renaming columns
new_uk_boys_names_df = new_uk_boys_names_df.rename(index=str, columns={"Rank": "rank", "Name": "name", "Count": "count"})
new_uk_boys_names_df.head()
```

```python
#Setting the index of the dataframe
new_uk_boys_names_df = new_uk_boys_names_df.set_index('rank')
new_uk_boys_names_df.head()
```

```python
#Removing the commas from the 'count' column
new_uk_boys_names_df["count"]=new_uk_boys_names_df["count"].str.replace(',', '')
```

```python
#Turning the strings in the 'count' column to numbers
new_uk_boys_names_df["count"]=pd.to_numeric(new_uk_boys_names_df['count'])
```

```python
#Printing the head of the dataframe
new_uk_boys_names_df.head()
```

```python
#locating the rows with 68 as rank
new_uk_boys_names_df.loc[68]
```

```python
#Creating path for csv
uk_girls_names_file = "../Datasets/UK_Girls_2011.csv"
```

```python
#Creating a file path for csv
us_boys_names_file = "../Datasets/US_Boys_ Names_Rank.csv"
```

```python
#Reading csv
us_boys_names_df = pd.read_csv(us_boys_names_file, encoding='utf-8')
us_boys_names_df.head()
```

```python
#Dropping empty cells from csv
us_boys_names_df= us_boys_names_df.dropna(how='any')
us_boys_names_df.head()
```

```python
#Renaming the column
us_boys_names_df = us_boys_names_df.rename(index=str, columns={"Rank": "rank", "Name": "name", "Number": "number"})
us_boys_names_df.head()
```

```python
#Setting the 'rank' column as index
us_boys_names_df = us_boys_names_df.set_index('rank')
us_boys_names_df.head()
```

```python
#Creating file path for csv
us_girls_names_file = "../Datasets/US_Girls_Names_ Rank.csv"
```

```python
us_girls_names_df = pd.read_csv(us_girls_names_file, encoding='utf-8')
us_girls_names_df.head()
```

```python
#Dropping empty cells from dataframe
us_girls_names_df= us_girls_names_df.dropna(how='any')
us_girls_names_df.head()
```

```python
#Renaming column names
us_girls_names_df = us_girls_names_df.rename(index=str, columns={"Rank": "rank", "Name": "name", "Number": "number"})
us_girls_names_df.head()
```

```python
#Setting the rank column as index
us_girls_names_df = us_girls_names_df.set_index('rank')
us_girls_names_df.head()
```

```python
#Creating a connection string and connecting to creating engine
rds_connection_string = "postgres:password:5432/Baby_Names_DB"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
#Checking table names in Postgres
engine.table_names()
```

```python
#Appending cleaned table to Postgres
new_uk_boys_names_df.to_sql('uk_boys_names', engine, if_exists='append')
```

```python
#Appending cleaned table to Postgres
new_uk_girls_names_df.to_sql('uk_girls_names', engine, if_exists='append')
```

```python
#Appending cleaned table to Postgres
us_girls_names_df.to_sql('us_girls_names', engine, if_exists='append')
```

```python
#Appending cleaned table to Postgres
us_boys_names_df.to_sql('us_boys_names', engine, if_exists='append')
```

```python
#Read table inside Postgres
pd.read_sql_query('select * from uk_boys_names', con=engine).head()
```

```python
#Read table inside Postgres
pd.read_sql_query('select * from uk_girls_names', con=engine).head()
```

```python
#Read table inside Postgres
pd.read_sql_query('select * from us_boys_names', con=engine).head()
```

```python
us_boys_names_file = "../Datasets/US_Boys_ Names_Rank.csv"
```

```python
us_boys_names_df = pd.read_csv(us_boys_names_file, encoding='utf-8')
us_boys_names_df.head()
```

```python
us_boys_names_df= us_boys_names_df.dropna(how='any')
us_boys_names_df.head()
```

```python
us_boys_names_df = us_boys_names_df.rename(index=str, columns={"Rank": "rank", "Name": "name", "Number": "number"})
us_boys_names_df.head()
```

```python
us_boys_names_df = us_boys_names_df.set_index('rank')
us_boys_names_df.head()
```

```python
us_girls_names_file = "../Datasets/US_Girls_Names_ Rank.csv"
```

```python
us_girls_names_df = pd.read_csv(us_girls_names_file, encoding='utf-8')
us_girls_names_df
```

```python
us_girls_names_df= us_girls_names_df.dropna(how='any')
us_girls_names_df.head()
```

```python
us_girls_names_df = us_girls_names_df.rename(index=str, columns={"Rank": "rank", "Name": "name", "Number": "number"})
us_girls_names_df.head()
```

```python
us_girls_names_df = us_girls_names_df.set_index('rank')
us_girls_names_df.head()
```

```python
rds_connection_string = "postgres:password:5432/Baby_Names_DB"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
new_uk_boys_names_df.to_sql('uk_boys_names', engine, if_exists='append')
```

```python
new_uk_girls_names_df.to_sql('uk_girls_names', engine, if_exists='append')
```

```python
us_girls_names_df.to_sql('us_girls_names', engine, if_exists='append')
```

```python
us_boys_names_df.to_sql('us_boys_names', engine, if_exists='append')
```
